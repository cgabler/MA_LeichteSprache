```python
for x in range(0, 10):
    df = pd.read_csv("../../02_Data/LS_mdr_gpt.csv")

    for i in range(0,100):
        index = 100*x + i

        messages = [ {"role": "system", "content": 
              "Ich werde dir einen Text in Leichter Sprache geben, übersetze mir diesen bitte in einen allgemeinen Text:"} ]
        text = df.at[index, 'text_leicht']
        if isinstance(text, str):
            if len(text) > 50:
                message = text
                if message:
                    messages.append(
                        {"role": "user", "content": message},
                    )

                    print(f"Starting: ", index, "(", df.at[index, 'title'], ") with Length: ", len(message))
                    try:
                        chat = openai.ChatCompletion.create(
                            model="gpt-3.5-turbo", messages=messages
                        )
                    except:
                        print("error at ", index)
                reply = chat.choices[0].message.content
                df.at[index, 'text_normal'] = reply
                
    
    print(f"##### WRITING ####")
    df.to_csv("../../02_Data/LS_mdr_gpt.csv")
```

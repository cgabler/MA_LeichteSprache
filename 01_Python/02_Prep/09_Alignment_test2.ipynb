```python
import pandas as pd

df = pd.read_csv('../02_Data/stuttgart_corrected.csv')
```

```python
df = df.dropna()
```

```python
df['Text_Leicht_lem'] = ""
df['Text_Allgemein_lem'] = ""
df['Sentences_Leicht_lem'] = ""
df['Sentences_Allgemein_lem'] = ""
df['Sentences_Leicht'] = ""
df['Sentences_Allgemein'] = ""
```

```python
for index, row in df.iterrows():
    doc_leicht = nlp(row["Text_Leicht"])
    doc_allg = nlp(row["Text_Allgemein"])
#    doc_leicht_lem = nlp(row["Text_Leicht_lem"])
#    doc_allg_lem = nlp(row["Text_Allgemein_lem"])
    df.at[index, "Sentences_Leicht"] = list(doc_leicht.sents)
    df.at[index, "Sentences_Allgemein"] = list(doc_allg.sents)
#    df.at[index, "Sentences_Leicht_lem"] = list(doc_leicht_lem.sents)
#    df.at[index, "Sentences_Allgemein_lem"] = list(doc_allg_lem.sents)
```

```python
lstSentences_leicht = []
lstSentences_allg = []
for index, row in df.iterrows():
    for i in range(len(df.at[index, "Sentences_Leicht"])):
        doc_leicht = nlp(str(df.at[index, "Sentences_Leicht"][i]))
        lstSentences_leicht.append(" ".join(token.lemma_ for token in doc_leicht))
        df.at[index, "Sentences_Leicht_lem"] = lstSentences_leicht

    for i in range(len(df.at[index, "Sentences_Allgemein"])):
        doc_allg = nlp(str(df.at[index, "Sentences_Allgemein"][i]))
        lstSentences_allg.append(" ".join(token.lemma_ for token in doc_allg))
        df.at[index, "Sentences_Allgemein_lem"] = lstSentences_allg
```

```python
def calculate_similarity(sentence1, sentence2):
    doc1 = nlp(sentence1)
    doc2 = nlp(sentence2)
    return doc1.similarity(doc2)
```

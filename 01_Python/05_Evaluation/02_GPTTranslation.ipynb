```python
import pandas as pd
import os
import openai
openai.api_key = ""
```

```python
df = pd.read_csv('../02_Data/GPT_testset.csv',index_col=[0])

df = df.reset_index()
df = df.drop(['index'], axis=1)
```

```python
message = ''
message += 'Wende dein eigenes Verständnis für Leichte Sprache an, ergänze und überschreibe falls notwendig mit folgenden konkreten Regeln. Nachfolgend erhältst du Texte in Allgemeiner Sprache die du anhand der Regeln übersetzt in leichte Sprache.'
message += 'Jede Regel und zugehöriges Beispiel ist in Triplequotes markiert.'
message += '"""Verwende simple Wörter die ein Grundschüler verstehen kann. (Beispiel: "erlauben" statt "genehmigen")"""'
message += '"""Benutze Wörter die etwas genau beschreiben (Beispiel: "Bus und Bahn" statt "Öffentlicher Nahverkehr", oder "Zeitungen" statt "Publikationen")"""'
message += '"""Verzichte auf Fach- und Fremdwörter, versuche deutsche leicht verständliche Worte zu finden. (Beispiel: "Arbeits-Gruppe" statt "Workshop")"""'
message += '"""Benutze immer die gleichen Wörter für gleiche Dinge (Beispiel: Verwende durchgehend das Wort "Tablette" und wechsle nicht zu "Pille")"""'
message += '"""Trenne lange Wörter mit einem Binde-Strich, trenne nach jedem Wort das in seiner Grundform allein stehen könnte. (Beispiel: "Bundes-Gleich-Stellungs-Gesetz" statt "Bundesgleichstellungsgesetz")"""'
message += '"""Benutze Verben statt Nomen (Beispiel: "Morgen wählen wir den Beirat" statt "Morgen ist die Wahl zum Beirat")"""'
message += '"""Vermeide den Genitiv, verwende "von dem", "vom" oder ähnliche Worte um es zu umgehen. Schreibe statt "Stefans Buch" lieber "das Buch von Stefan" """'
message += '"""Benutze positive Sprache, versuche Worte wie "nein", "kein", "nicht" die zur Verneinung eines Umstandes verwendet werden zu ersetzen. (Beispiel: "Peter ist gesund" statt "Peter ist nicht krank")"""'
message += '"""Verwende arabische Zahlen anstelle von ausgeschriebenen Zahlen. (Beispiel: "5" statt "fünf")"""'
message += '"""Vermeide alte Jahreszahlen, verwende 100 Jahre als Faustregel. (Beispiel: "vor mehr als 100 Jahren" statt "1867")"""'
message += '"""Vermeide hohe Zahlen, verwende 500 als Obergrenze ab der du Umschreibungen verwendest. (Beispiel: "viele Menschen" statt "14576 Menschen")"""'
message += '"""Schreibe kurze Sätze, die Satzlänge ist auf 8 Worte limitiert."""'
message += '"""Schreibe Sätze mit nur einer Aussage pro Satz (Beispiel: "Ich kann Ihnen helfen. Bitte sagen Sie mir: Was wünschen Sie?" statt "Wenn Sie mir sagen, was Sie wünschen, kann ich Ihnen helfen.")"""'
message += '"""Wenn du Sätze trennst darfst du folgende Sätze mit "Und", "Oder", "Aber" beginnen"'
message += '"""Benutze einfachen Satzbau in der Form von Subjekt - Prädikat - Objekt (Beispiel: "Wir fahren zusammen in den Urlaub" statt "Zusammen fahren wir in den Urlaub.")"""'
message += '"""Vermeide Fragen im Text, entferne rhetorische Fragen, präsentiere den Leser mit direkten Aussagen."""'
```

```python
#index = 1
for index in range(20,40):
    messages = [ {"role": "system", "content": 
                message} ]

    messages.append(
        {"role": "user", "content": df.iloc[index]['text_normal'].strip()},
    )
    chat = openai.ChatCompletion.create(
        model="gpt-4", messages=messages
    )
    print(f"Result: ", chat.choices[0].message.content)
    df.at[index, 'text_translation'] = chat.choices[0].message.content
```

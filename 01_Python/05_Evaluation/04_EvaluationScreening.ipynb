```python
#index = 1
for index in range(0,30):
   df_gpt.at[index, 'hurraki_difference'] = df_gpt.at[index, 'hurraki_translation'] - df_gpt.at[index, 'hurraki_original']
   df_gpt.at[index,'length_original'] = len(df_gpt.at[index, 'text_leicht'])
   df_gpt.at[index,'length_translation'] = len(df_gpt.at[index, 'text_translation'])
   df_gpt.at[index,'length_standard'] = len(df_gpt.at[index, 'text_normal'])
   df_gpt.at[index,'error_original'] = df_gpt.at[index,'hurraki_original'] / df_gpt.at[index,'length_original']  
   df_gpt.at[index,'error_translation'] = df_gpt.at[index,'hurraki_translation'] / df_gpt.at[index,'length_translation']
```

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
palette = sns.color_palette("vlag", as_cmap=True)
print(plt.style.available)
# Make default density plot
#sns.kdeplot(data=data,x='x',y='y')
sns.lineplot(data=df_gpt, y=df_gpt['error_original'], x=df_gpt.index, label='Referenztexte', palette=palette)
sns.lineplot(data=df_gpt, y=df_gpt['error_translation'], x=df_gpt.index, label='Übersetzte Texte', palette=palette)

plt.title('Fehlerverhältnis')
plt.xlabel('Index')
plt.ylabel('Verteilung')
plt.legend(loc="upper left")
plt.style.use('seaborn-v0_8-colorblind')
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(df_gpt['error_original'])
plt.plot(df_gpt['error_translation'])
plt.title('Vergleich Hurraki Fehler')
plt.ylabel('Fehlerverhältnis')
plt.xlabel('Text #')
plt.legend(loc="upper right", labels=['Referenztext', 'Übersetzung'])
plt.show()
```

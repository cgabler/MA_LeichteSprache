```python
import pandas as pd

df = pd.read_csv("../02_Data/LS_einfachstars_gpt.csv")
try:
    df = df.drop('Unnamed: 0', axis=1)
except:
    print("error deleting unnamed")
```

```python
print('LÃ¤nge einfachstars. ', len(df))
```

```python
df['text_normal'] = ""
df['text_normal_sents'] = ""
```

```python
for x in range(0, 8):    
    df_copy = pd.DataFrame(columns=['URL', 'title', 'text_leicht', 'source', 'token', 'sents', 'text_leicht_sents', 'text_normal', 'text_normal_sents'])
    for i in range(0, 1000):
        index = 1000*x + i
        try:
            df_copy.loc[i] = df.loc[index]
        except:
            print('end')
    path = '../02_Data/LS_einfachstars_gpt'+str(x)+'.csv'
    df_copy.to_csv(path)
```

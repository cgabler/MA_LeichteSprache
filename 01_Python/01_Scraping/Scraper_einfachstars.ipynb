```python
import requests
import pandas as pd
from bs4 import BeautifulSoup, NavigableString
```

```python
df = pd.DataFrame(columns=['URL', 'Titel', 'Text_Leicht'])
i = 1

for x in range(1, 744):
    URL = "https://einfachstars.info/blog/index.html?page=" + str(x)
    print(URL)
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")

    articles = soup.find_all("article", class_="post")
    for a in articles:
        url = ''
        title = ''
        text = ''
        href = a.find_all("a")
        for u in href:
            if u.has_attr('href'):
                if u['href'].startswith('/blog') and "#respond" not in u['href']:
                    url = u['href']
                    title = u.getText()
        
        print('     '+title)
        
        lst_p = a.find_all("p")
        for p in lst_p:
            if p.has_attr('class'):
                if "image_caption" not in p['class'] and "image-source" not in p['class']:
                    text += p.getText(separator=u' ')   
            else:
                text += p.getText(separator=u' ')   
        

        df_article = pd.DataFrame({
            'URL': url,
            'Titel': title,
            'Text_Leicht': text},
            index=[i]
        )

        i += 1
        df = pd.concat([df, df_article])
```

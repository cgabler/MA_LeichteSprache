```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import unicodedata

##tabs-box4
URL_Sachsen = "https://www.mdr.de/nachrichten-leicht/rueckblick/leichte-sprache-rueckblick-buendelgruppe-sachsen-100.html"
URL_Thueringen = "https://www.mdr.de/nachrichten-leicht/rueckblick/leichte-sprache-rueckblick-buendelgruppe-thueringen-100.html"
URL_SachsenAnhalt = "https://www.mdr.de/nachrichten-leicht/rueckblick/leichte-sprache-rueckblick-buendelgruppe-sachsen-anhalt-100.html"
page = requests.get(URL_Sachsen)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
df = pd.DataFrame()
df['URL'] = ""
df['title'] = ""
df['text_leicht'] = ""
```

```python
_index = 0

for i in range(1,4):
    URL = URL_Sachsen + '#tabs-box'+str(i)
    page = requests.get(URL_Sachsen)
    soup = BeautifulSoup(page.content, "html.parser")
    teaser = soup.find_all("div", {"class": "teaser"})
    for t in teaser:
        links = t.find_all("a", {"class": "linkAll"})
        for l in links:
            url = "https://www.mdr.de" + l["href"]
            print(url)
            df = pd.concat([df, pd.DataFrame({
                'URL': url,
                'title': "",
                'text_leicht': ""},
                index=[_index]
            )])
            _index += 1
        
```

```python

for i in range(1,4):
    URL = URL_SachsenAnhalt + '#tabs-box'+str(i)
    page = requests.get(URL_SachsenAnhalt)
    soup = BeautifulSoup(page.content, "html.parser")
    teaser = soup.find_all("div", {"class": "teaser"})
    for t in teaser:
        links = t.find_all("a", {"class": "linkAll"})
        for l in links:
            url = "https://www.mdr.de" + l["href"]
            print(url)
            df = pd.concat([df, pd.DataFrame({
                'URL': url,
                'title': "",
                'text_leicht': ""},
                index=[_index]
            )])
            _index += 1
```

```python
for index, row in df.iterrows():
    URL = df.at[index, "URL"]
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    title = soup.find("span", {"class": "headline"}).text
    text = ""
    for p in soup.find_all("p", {"class": "text"}):
        text += p.text

    df.at[index, "title"] = title
    df.at[index, "text_leicht"] = text.strip()

```

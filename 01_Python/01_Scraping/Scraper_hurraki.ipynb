```python
import requests
import pandas as pd
from bs4 import BeautifulSoup, NavigableString

URL = "https://hurraki.de/wiki/Hurraki:Artikel_von_A_bis_Z"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
links = soup.find_all("a")
links
```

```python
lst = []
for l in links:
    if l.has_attr('href'):
        if l['href'].startswith('/wiki') and "rter_von_A_bis_Z" not in l['href']:
            lst.append("https://hurraki.de"+l['href'])
        
lst
```

```python
df = pd.DataFrame(columns=['URL', 'Titel', 'Text_Leicht'])
i = 1

for url in lst:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    title
    text = ''
    bln_stop = False
    for t in soup.find_all("span", class_="mw-page-title-main"):
        title = t.getText()

        try:
            a = soup.find("span", {"id":"Genaue_Erklärung"})
            for p in a.parent.next_siblings:
                if "Hilf mit! Prüfe den Text auf" in p.getText():
                    bln_stop = True
                else:
                    text += p.getText()
                if bln_stop:
                    break
        except Exception as error:
            print("Error with ", title, url)
            print(error)

    df_article = pd.DataFrame({
        'URL': url,
        'Titel': title,
        'Text_Leicht': text},
        index=[i]
    )
    
    i += 1
    df = pd.concat([df, df_article])
            
df        
```

```python
df.to_csv("../../02_Data/hurraki.csv")
```

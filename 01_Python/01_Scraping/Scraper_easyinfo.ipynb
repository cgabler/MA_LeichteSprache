```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup, NavigableString
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# Define the URL
url = "https://infoeasy-news.ch/"

# load the web page
driver.get(url)

# set maximum time to load the web page in seconds
driver.implicitly_wait(10)
```

```python
SCROLL_PAUSE_TIME = 3

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
links = soup.find_all("a")
l=[]
for a in links:
    if not a.has_attr('class') and "Webmaster" not in a.getText() and "Administrator" not in a.getText():
        l.append(a['href'])
```

```python
df = pd.DataFrame(columns=['URL', 'Titel', 'Text_Leicht'])
i = 1

for url in l:
    if url not in "http://infoeasy-news.ch/wp-content/uploads/speaker/post-3001.mp3?cb=1655544699.mp3":
        print(url)
        page = requests.get(url)
        soup = BeautifulSoup(page.content, "html.parser")

        title = ''
        text = ''
        for h in soup.find_all('h2'):
            if not h.has_attr('class'):
                title = h.getText()

        for div in soup.find_all("div", class_="fusion-text"):
            text += div.getText(separator=u' ')

        df_article = pd.DataFrame({
            'URL': url,
            'Titel': title,
            'Text_Leicht': text},
            index=[i]
        )

        i += 1
        df = pd.concat([df, df_article])
```
